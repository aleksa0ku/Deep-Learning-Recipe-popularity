### Handle ingredients and instructions lists

```python
# Extract and flatten ingredients and instructions, and count them
def flatten_ingredients_and_count(ingredients_list):
    if isinstance(ingredients_list, list):
        flattened = []
        count = 0
        for category in ingredients_list:
            if 'ingredients' in category:
                for ingredient in category['ingredients']:
                    primary_unit = ingredient.get('primary_unit') or {}
                    quantity = primary_unit.get('quantity', '')
                    unit = primary_unit.get('display', '')
                    name = ingredient.get('name', '')
                    extra_comment = ingredient.get('extra_comment', '')
                    parts = [quantity, unit, name, extra_comment]
                    parts = [part for part in parts if part]  # Remove empty strings
                    flattened.append(' '.join(parts))
                    count += 1
        return '; '.join(flattened), count
    return '', 0

def flatten_instructions_and_count(instructions_list):
    if isinstance(instructions_list, list):
        steps = [instruction['display_text'] for instruction in instructions_list]
        return ' '.join(steps), len(steps)
    return '', 0
```

```python
# Apply the functions to create new columns
df_merge['flattened_ingredients'], df_merge['num_ingredients'] = zip(*df_merge['ingredient_sections'].apply(flatten_ingredients_and_count))
df_merge['flattened_instructions'], df_merge['num_steps'] = zip(*df_merge['instructions'].apply(flatten_instructions_and_count))
```

Given the importance of the `score` colum I will use a more advanced imputation method KNN imputation. This method uses the k-nearest neighbors to impute missing values based on the similarity of other feature values. The neighbors are the k most similar instances (rows) to the instance with the missing value.

```python
from fancyimpute import KNN
import numpy as np

# Replace zero values in 'score' with NaN for imputation
df_numerical['score'].replace(0, np.nan, inplace=True)

# Select the columns for KNN imputation
columns_to_impute = ['score']

# Include all relevant features for imputation
features_for_imputation = [col for col in df_numerical.columns if col not in columns_to_impute]

# Combine the columns to impute with features
df_for_imputation = df_numerical[columns_to_impute + features_for_imputation]

# Apply KNN imputation
df_imputed = pd.DataFrame(KNN(k=5).fit_transform(df_for_imputation), columns=columns_to_impute + features_for_imputation)

# Update the original df_numerical with imputed values
df_numerical.update(df_imputed[columns_to_impute])

# Verify the changes
print(df_numerical['score'].describe())
```

### Handle outliers

Outliers can significantly affect the performance of machine learning models. Many algorithms, particularly those based on distance metrics (e.g., K-Nearest Neighbors, Support Vector Machines) and linear models (e.g., Linear Regression, Logistic Regression), are sensitive to outliers. These algorithms can produce biased or skewed results when trained on data containing outliers.

Bias in Linear Models: In linear regression, outliers can disproportionately affect the slope of the line of best fit, leading to inaccurate predictions.

Distance-Based Algorithms: Outliers can distort the distance calculations, leading to incorrect classifications or cluster formations.

Capping: Cap the extreme values at a certain percentile. This reduces their influence without removing them, which would cause data loss.

1. **Visualize Outliers for All Columns**: Box plots are used to visualize the presence of outliers in all numeric columns.
2. **Identify Outliers Using IQR**: The IQR method is used to identify outliers for each column. The lower bound is calculated as Q1 - 1.5 * IQR, and the upper bound is calculated as Q3 + 1.5 * IQR.
3. **Handle Outliers by Capping**: Outliers are capped to the lower and upper bounds for each numeric column. This means values below the lower bound are set to the lower bound, and values above the upper bound are set to the upper bound.
4. **Save the Updated DataFrame**: The cleaned dataframe with outliers handled is saved back to a pickle file for future use.

This comprehensive approach ensures that outliers are handled across all relevant features in the dataset, improving the overall quality of the data for model training.

```python
# List of columns to visualize
selected_columns = [
    'cook_time', 'prep_time', 'total_time', 
    'ratings_negative', 'ratings_positive', 'score',
    'protein', 'fat', 'calories', 'sugar', 'carbohydrates', 'fiber'
]

# Filter the DataFrame to include only the selected columns
df_selected = df_numerical[selected_columns]
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(15, 20))

# Create box plots for each selected column
for i, col in enumerate(selected_columns, 1):
    plt.subplot(len(selected_columns) // 3 + 1, 3, i)
    sns.boxplot(x=df_selected[col])
    plt.title(f'Box plot of {col}')

plt.tight_layout()
plt.show()
```

```python
# Identify and handle outliers using IQR method for all numeric columns
def cap_outliers(df, cols, threshold=1.5):
    """
    Cap outliers in the dataframe using the IQR method.
    
    Args:
    df (pd.DataFrame): DataFrame containing the data.
    cols (list): List of columns to cap outliers in.
    threshold (float): The multiplier for the IQR to define outliers.
    
    Returns:
    pd.DataFrame: DataFrame with outliers capped.
    """
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR
        df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
    return df


# Cap outliers in the selected columns
df_numerical = cap_outliers(df_numerical, selected_columns)

# Verify the changes
df_numerical[selected_columns].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(15, 20))

# Create box plots for each selected column after capping outliers
for i, col in enumerate(selected_columns, 1):
    plt.subplot(len(selected_columns) // 3 + 1, 3, i)
    sns.boxplot(x=df_numerical[col])
    plt.title(f'Box plot of {col} after capping outliers')

plt.tight_layout()
plt.show()
```

The box plots show that the outliers have been effectively capped, resulting in more compact and controlled distributions for each column. The data is now better prepared for model training.

```python
# Function to map TF-IDF feature indices to actual terms
def map_tfidf_features(feature_names, prefix):
    return {f'{prefix}{i}': term for i, term in enumerate(feature_names)}

# Assuming TF-IDF feature names in your DataFrame are prefixed with the text feature names
name_prefix = 'name_tfidf_'
ingredients_prefix = 'flattened_ingredients_tfidf_'
instructions_prefix = 'flattened_instructions_tfidf_'
keywords_prefix = 'keywords_tfidf_'

# Map TF-IDF features to terms
name_tfidf_features = map_tfidf_features(name_feature_names, name_prefix)
ingredients_tfidf_features = map_tfidf_features(ingredients_feature_names, ingredients_prefix)
instructions_tfidf_features = map_tfidf_features(instructions_feature_names, instructions_prefix)
keywords_tfidf_features = map_tfidf_features(keywords_feature_names, keywords_prefix)

# Combine all mappings into a single dictionary
tfidf_mappings = {**name_tfidf_features, **ingredients_tfidf_features, **instructions_tfidf_features, **keywords_tfidf_features}
```

```python
# Compute correlation of each feature with the target variable 'score'
correlation_with_target = df_numerical.corr()['score'].sort_values(ascending=False)
```

```python
# Select top 20 features based on absolute correlation with 'score'
top_features = correlation_with_target.abs().sort_values(ascending=False).head(21).index.tolist()
top_features.remove('score')  # Remove 'score' from the list of features
print("Top features selected based on correlation:\n", top_features)
```

```python
# Compute the correlation matrix for the top features using original names
corr_matrix_top_features = df_numerical[top_features].corr()

# Rename the columns and index for the correlation matrix using actual terms
top_features_renamed = [tfidf_mappings.get(feature, feature) for feature in top_features]
corr_matrix_top_features.columns = top_features_renamed
corr_matrix_top_features.index = top_features_renamed

# Plot the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix_top_features, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix with Actual Terms')
plt.show()
```

```python
# Define a custom dataset class
class RecipeDataset(Dataset):
    def __init__(self, df, tokenizer, max_length):
        self.df = df
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        text = " ".join([str(row['keywords']), str(row['name']), str(row['flattened_ingredients']), str(row['flattened_instructions'])])
        inputs = self.tokenizer(text, padding='max_length', truncation=True, max_length=self.max_length, return_tensors="pt")
        inputs = {key: val.squeeze(0) for key, val in inputs.items()}
        numerical_features = torch.tensor(row[['cook_time', 'prep_time', 'total_time', 'ratings_negative', 'ratings_positive', 'protein', 'fat', 'calories', 'sugar', 'carbohydrates', 'fiber', 'num_ingredients', 'num_steps']].values, dtype=torch.float)
        label = torch.tensor(row['score'], dtype=torch.float)
        return inputs, numerical_features, label
```

```python
# Initialize the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Split the data
train_df, test_df = train_test_split(df_combined, test_size=0.2, random_state=42)
```

```python
# Create DataLoader instances
train_dataset = RecipeDataset(train_df, tokenizer, max_length=512)
test_dataset = RecipeDataset(test_df, tokenizer, max_length=512)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)
```

```python
# Define the combined model
class CombinedModel(nn.Module):
    def __init__(self, bert_model, numerical_feature_size):
        super(CombinedModel, self).__init__()
        self.bert = bert_model
        self.dropout = nn.Dropout(0.3)
        self.fc1 = nn.Linear(768 + numerical_feature_size, 128)
        self.fc2 = nn.Linear(128, 1)

    def forward(self, input_ids, attention_mask, numerical_features):
        bert_output = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = bert_output.last_hidden_state[:, 0, :]  # CLS token
        combined_input = torch.cat((cls_output, numerical_features), dim=1)
        x = self.dropout(combined_input)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
# Define the training loop
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CombinedModel(bert_model, numerical_feature_size=train_df[['cook_time', 'prep_time', 'total_time', 'ratings_negative', 'ratings_positive', 'protein', 'fat', 'calories', 'sugar', 'carbohydrates', 'fiber', 'num_ingredients', 'num_steps']].shape[1]).to(device)
optimizer = optim.AdamW(model.parameters(), lr=2e-5)
criterion = nn.MSELoss()
```

```python
def train_epoch(model, dataloader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for batch in dataloader:
        optimizer.zero_grad()
        inputs, numerical_features, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        numerical_features = numerical_features.to(device)
        labels = labels.to(device).unsqueeze(1)
        outputs = model(inputs['input_ids'], inputs['attention_mask'], numerical_features)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)
```

```python
def evaluate(model, dataloader, device):
    model.eval()
    predictions, true_labels = [], []
    with torch.no_grad():
        for batch in dataloader:
            inputs, numerical_features, labels = batch
            inputs = {key: val.to(device) for key, val in inputs.items()}
            numerical_features = numerical_features.to(device)
            labels = labels.to(device).unsqueeze(1)
            outputs = model(inputs['input_ids'], inputs['attention_mask'], numerical_features)
            predictions.extend(outputs.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())
    return predictions, true_labels
```

```python
# Train the model
num_epochs = 3
for epoch in range(num_epochs):
    train_loss = train_epoch(model, train_loader, optimizer, criterion, device)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss}")
```
